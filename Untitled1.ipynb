```python
import pandas as pd
```

```python
#punto 1
def leer_precios(filename):
    for k in range(10):
        df = pd.read_excel(filename,
                           skiprows = k,
                           usecols = list(range(25)))
        if df.iloc[0,0] == 'Fecha':
            df = pd.read_excel(filename,
                               skiprows = k+1,
                               usecols = list(range(25)))
            break
        
    return(df)
    
```

```python
#filenames = !ls precios/
#dfs = []
#for filename in filenames:
#     dfs.append(leer_precios("Precios/"+filename))

filenames = !ls precios/*.xls*
filenames

dfs = []
for filename in filenames:
    dfs.append(leer_precios(filename))

dfs = pd.concat(dfs)
dfs.size

dfs1 = pd.melt(dfs, id_vars='Fecha')
dfs1
```
